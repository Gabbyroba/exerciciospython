```python
n1=float(input('Digite o valor do limite total: '))
n2=float(input('Digite o valor da fatura atual: '))
n3=float(input('Digite o valor das parcelas futuras: '))
print(f'Limite total{n1} menos valor da fatua atual {n2} e valor das parcelas futuras {n3} é igual a Limite Disponível {(n1-n2-n3)}.')
```

```python
r=float(input('Quanto você tem na carteira? R$ '))
print(f'Com R${r} você pode comprar U$${(r/5.19):.2f}.')
```

```python
larg=float(input('Qual é a largura da parede? ')) 
altu=float(input('Qual é a altura da parede? '))
print(f'Sua parede tem a área de {(larg*altu):.2f}m²')
print(f'Para pintar sua parede é necessário um total de {(larg*altu)/2:.2f}l de tinta.')
```

```python
pre=float(input('Qual é o preço do produto? R$ '))
print(f'Com 5% de desconto o produto fica R${pre-(pre*5/100):.2f}.')
```

```python
sal=float(input('Qual é o salário atual do funcionário? R$ '))
print(f'Com o aumento de 15% o funcionário passa a receber R$ {sal+(sal*15/100)}')
```
